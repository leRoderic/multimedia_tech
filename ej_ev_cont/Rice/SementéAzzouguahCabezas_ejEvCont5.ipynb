```python
from math import ceil, log2

def rice_codification(N, M):
    ret, signo, Q, R = "", "", "", ""
    # Comprobamos signo
    signo = "0" if N < 0 else "1"
    # Calculamos el cociente
    Q = "1" * int(abs(N)/M) + "0"
    # Calculamos el residuo
    R = "{0:b}".format(abs(N)%M).zfill(ceil(log2(M)))
    return signo + Q + R
```

##### (a) ¿Cuántos bits son necesarios para codificar todos los enteros entre -1023 y +1023 (en codificación binaria natural con bit de signo)?

Son necesarios 10 bits para representar de 0 a 1023 enteros, más 1 bit de signo -> total 11 bits.

##### (b) Calculad el código Rice de todos los enteros N entre -1023 y +1023 con M = 32.

```python
for i in range(-1023, 1024):
    print("Int: {}  Coded: {}".format(i, rice_codification(i, 32)))
```

##### (c) ¿Para qué rango de valores de N, el código Rice requiere menos bits que la codificación binaria natural con bit de signo? Si la mayoría de datos a codificar pertenecen a este rango (rango de entrada óptima), el uso de código Rice será aconsejable (habrá ahorro de bits) ¿Cuál es el máximo ahorro de bits?

```python
def test_stuff(rangeStart, rangeEnd, M):
    for i in range(rangeStart, rangeEnd):
        riceLen, lenBin = len(rice_codification(i, M)), len("{0:b}".format(i).zfill(ceil(log2(rangeEnd)))) + 1
        if riceLen > lenBin - 1:
            print("\n\nRange found between {} and {}".format(rangeStart, i - 1))
            break
        print("Int: {} \tRice bits: {} \tBinary bits: {}".format(i, riceLen, lenBin))

test_stuff(0, 685, 32)
```

Teniendo M=32, hemos encontrado para los enteros compredidos en el rango [0, 127] se usan menos bits con la codificación Rice que en binario natural.

##### (d) ¿Para qué valor de M el rango de entrada óptima se extiende entre -255 y +255? ¿Cuál es el máximo ahorro de bits ahora?

```python
for i in range(9520):
    test_stuff(-255, 255, i + 1)
```
