```python
import numpy as np
import math
def riceCodification(N, M):
    ret, signo, Q, R = "", "", "", ""
    # Comprovamos signo
    signo = "0" if N < 0 else "1"
    #Calculamos el cociente
    Q = "1" * int(abs(N)/M) + "0"
    #Calculamos el residuo
    R = "{0:b}".format(abs(N)%M).zfill(math.ceil(np.log2(M)))

    ret = signo + Q + R
    
    return ret
```

##### (a) ¿Cuántos bits son necesarios para codificar todos los enteros entre -1023 y +1023 (en codificación binaria natural con bit de signo)?

10 bits para representar de 0 a 1023 más 1 bit de sino -> 11 bits

##### (b) Calculad el código Rice de todos los enteros N entre -1023 y +1023 con M = 32.

```python
for i in range(-1023, 1024):
    riceC = riceCodification(i,32)
    print(i, "Codificación:", riceC)
```

##### (c) ¿Para qué rango de valores de N, el código Rice requiere menos bits que la codificación binaria natural con bit de signo? Si la mayoría de datos a codificar pertenecen a este rango (rango de entrada óptima), el uso de código Rice será aconsejable (habrá ahorro de bits) ¿Cuál es el máximo ahorro de bits?

```python
for i in range(1024):
    if i < 0:
        #binario + bit de signo
        binNormal = len(bin(i)) - 3 + 1
    else:
        binNormal = len(bin(i)) - 2 + 1
    riceC = riceCodification(i,32)
    print(i, "Codificación Rice:", binNormal, " Binario:", len(riceC))
    #Si la codificación Rice es mas óprima:
    if len(riceC) <= binNormal:
        print(i ,"es un valor óprimo a representar con codificación Rice")
```

Tras realizar las respectivas pruebas, hemos podido observar que para la ventana indicada i M=32, es más óptimo utilizar codificación binaria con bit de signo que codificación Rice
